```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
sns.set()
```

```python
filenameCARS = r'C:\Users\wilso\Downloads\bar_chart_data.csv'
df_used_cars = pd.read_csv(filenameCARS, delimiter = ',')
df_used_cars
```

```python
plt.figure(figsize= (9,6))

plt.bar(x = df_used_cars['Brand'],
        height = df_used_cars['Cars Listings'],
        color = "midnightblue")

plt.xticks(rotation = 45, fontsize = 13)
plt.yticks(fontsize = 13)
plt.title('Cars Listings by Brand', fontsize = 16, fontweight = "bold")
plt.ylabel("Number of Listings", fontsize = 13)

plt.show()
plt.savefig("Used Cars Bar.png")
```

```python
filenameENGINE = r'C:\Users\wilso\Downloads\stacked_area_chart_data.csv'
df_engine_type = pd.read_csv(filenameENGINE, delimiter = ',')
df_engine_type.head()
```

```python
sns.set_style('white')

plt.figure(figsize= (15, 6))
colors = ["#011638", '#7e2987', '#ef2026']
plt.stackplot(df_engine_type['Year'],
              df_engine_type['Diesel'],
              df_engine_type['Petrol'],
              df_engine_type['Gas'],
              colors = colors,
              edgecolor = 'None'
              )
plt.legend(["Diesel", 'Petrol', 'Gas'],
           loc = 'upper left')
plt.ylabel('Number of Cars', fontsize = 15, weight = 'bold')
plt.xticks(df_engine_type["Year"], rotation = 45)
sns.despine()
plt.show()

plt.savefig('Number of Cars - Stacked Area Chart')
```

```python
filenameREAL_ESTATE = r'C:\Users\wilso\Downloads\histogram_data.csv'
df_real_estate = pd.read_csv(filenameREAL_ESTATE, delimiter = ',')
df_real_estate
```

```python
sns.set_style('white')
plt.figure(figsize = (8,6))
plt.hist(df_real_estate['Price'], 
         bins= 8,
        color = '#108A99')

plt.title('Distribution of Real Estate Prices', fontsize = 19, fontweight = 'bold')
plt.xlabel("Price in (000' $)", fontsize = 14, fontweight = 'bold')
plt.ylabel('Number of Properties', fontsize = 14, fontweight = 'bold')

sns.despine()

plt.show()
```

```python
plt.figure(figsize = (12, 8))

scatter = plt.scatter(df_real_estate['Area (ft.)'],
            df_real_estate['Price'],
            alpha= 0.6,
            c = df_real_estate['Building Type'], 
            cmap= 'viridis')

plt.title('Relationship between Area and Price of California Real Estate', 
          fontsize = 14,
          fontweight = 'bold')
plt.ylabel('Price in Thousand of Dollars', fontweight = 'bold')
plt.xlabel('Area (sq. ft)', fontweight = 'bold')

plt.legend(*scatter.legend_elements(), 
           loc = 'upper left', 
           title = "Building Type")

plt.show()
```

```python
plt.figure(figsize = (12,8))
sns.set_style('darkgrid')
sns.scatterplot(x = df_real_estate['Price'],
                y = df_real_estate['Area (ft.)'],
                hue=df_real_estate['Building Type'],
                palette=['black', 'darkblue', 'purple', 'pink', 'white'],
                s = 100)

plt.title('Relationship between Area and Price of California Real Estate', 
          fontsize = 14,
          fontweight = 'bold')
plt.ylabel('Price in Thousand of Dollars', fontweight = 'bold')
plt.xlabel('Area (sq. ft)', fontweight = 'bold')

plt.show()
```

```python
filenameAD_EXPENDITURE = r'C:\Users\wilso\Downloads\scatter_plot_ii.csv'
df_ad_expenditure = pd.read_csv(filenameAD_EXPENDITURE, delimiter = ',')
df_ad_expenditure
```

```python
# plt.figure(figsize = (10,8))
sns.set(rc = {'figure.figsize': (10,6)})
sns.regplot(x = 'Budget',
            y = 'Sales',
            data = df_ad_expenditure,
            #color = 'grey',
            scatter_kws= {'color': 'k'},
            line_kws= {'color': 'red'})

plt.title('Effect of Ad Expenditure on Sales', fontsize = 20, fontweight = 'bold')
plt.ylabel("Ad Expenditure in (000's $)", fontsize = 14, fontweight = 'bold')
plt.xlabel("Sales in (000's $)", fontsize = 14, fontweight = 'bold')


plt.show()
```

```python
# Linear regression plot, can oly support x and y in the form of strings or tidy data
sns.lmplot(x='Budget', 
           y='Sales', 
           data=df_ad_expenditure, 
           height = 10,
           scatter_kws= {'color': 'red'},
           line_kws= {'color': 'midnightblue'})

plt.title('Effect of Ad Expenditure on Sales', fontsize = 20, fontweight = 'bold')
plt.ylabel("Ad Expenditure in (000's $)", fontsize = 14, fontweight = 'bold')
plt.xlabel("Sales in (000's $)", fontsize = 14, fontweight = 'bold')

plt.show()
```

```python
from matplotlib.ticker import PercentFormatter
```

```python
filenameKD_Nuggets = r'C:\Users\wilso\Downloads\bar_line_chart_data.csv'
df_kdnuggets = pd.read_csv(filenameKD_Nuggets, delimiter = ',')
df_kdnuggets
```

```python
sns.set_style('white')

figure, ax = plt.subplots(figsize = (10,7))
ax.bar(df_kdnuggets['Year'],
       df_kdnuggets['Participants'],
       color = 'k')
ax.set_ylabel('Number of Participants', fontweight = 'bold')
ax.tick_params(axis = 'y', width = 2, labelsize = 'large')

ax1 = ax.twinx()
ax1.set_ylim(0, 1)
ax1.yaxis.set_major_formatter(PercentFormatter(xmax = 1.0))
ax1.plot(df_kdnuggets['Year'],
         df_kdnuggets['Python Users'],
         color = '#b60000',
         marker = 'D')
ax1.set_ylabel('Python Users', color = '#b60000', fontweight = 'bold')
ax1.tick_params(axis = 'y', colors = '#b60000', width = 2, labelsize = 'large')

ax.set_title('KD Nuggets Survey Python Users (2012 - 2019)', fontsize = 20, fontweight = 'bold')
```
